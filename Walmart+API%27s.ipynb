# Extracting Data From Walmart Open API

## Walmart Open API Tutorial using Python

<p>Walmart Open API(link-https://developer.walmartlabs.com) provides access to extensive product catalog. 
People need to register themselves in order to use API's. 
We need 'keys' in order to look up the feautres of each product. 
Walmart Open API is still in Beta Phase</p>

## The following API's are provided by Walmart Open API..

# Lookup API

## Provides price, availability etc of the item. Require two arguments: item id (required) and format (json or xml)

```python
#import packages:
import requests

#assign the api link to url:
url = 'http://api.walmartlabs.com/v1/items/12417832?format=json&apiKey=j4bcsts3aqkv2rc9fgacch8v'

#package the request, send the request, get response:
r = requests.get(url)

#decode the json to dict:
json_date = r.json()

#loop over json_data to print key,value pairs:
for k in json_date.keys():
    print(k + ':', json_date[k])
```
